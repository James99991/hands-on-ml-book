## The Dataset

### Source
This dataset is a modified version of the California Housing dataset available from [Luís Torgo's page](http://www.dcc.fc.up.pt/~ltorgo/Regression/cal_housing.html) (University of Porto). Luís Torgo obtained it from the StatLib repository (which is closed now). The dataset may also be downloaded from StatLib mirrors.

This dataset appeared in a 1997 paper titled *Sparse Spatial Autoregressions* by Pace, R. Kelley and Ronald Barry, published in the *Statistics and Probability Letters* journal. They built it using the 1990 California census data. It contains one row per census block group. A block group is the smallest geographical unit for which the U.S. Census Bureau publishes sample data (a block group typically has a population of 600 to 3,000 people).

### Tweaks
The dataset in this directory is almost identical to the original, with two differences:

* 207 values were randomly removed from the `total_bedrooms` column, so we can discuss what to do with missing data.
* An additional categorical attribute called `ocean_proximity` was added, indicating (very roughly) whether each block group is near the ocean, near the Bay area, inland or on an island. This allows discussing what to do with categorical data.

Note that the block groups are called "districts" in the Jupyter notebooks, simply because in some contexts the name "block group" was confusing.

### Performance Measure

We can go for RMSE or MAP (mean absolute error). RMSE is more common and penalises outliers more. If there were several outlier districts, MAP may be best.

- Distance measures can also be refered to as *norms*
- RMSE: Computing the root of sum of squares corresponds too the *Euclidian Norm*. Normal distance we are familair with.
- MAP: Computing the sum of absolute corresponds to the *l1* norm, also call *Manhattan Norm* - Measures distance between points in City if you could only travel along city blocks

```python
from __future__ import division, print_function, unicode_literals

import os
import pandas as pd
import tarfile
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = "datasets"
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

# NOTE: The HOUSING_URL is the URL that will prompt one to download the TAR file
# This is not to be confused with the location of the file through the git repository

# Function to download TAR files, save in specified location and extract them
# TAR files can contain multiple files

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    # Check if the folder doesn't exist in working folder
    # If it doesn't one will be created
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    # Path of file   
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

# Get data
fetch_housing_data()
```

## Creating Test Set

**Have to create test set before we proceed**

- Don't want to find patterns in test set that would influence decisions later. Models in production would thus not perform as well as expected - *Data Snooping Bias*

```python
import numpy

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data)) # np.random.seed(42) - to make things reproduceable
    test_set_size = int(len(data) * test_ratio) # number of rows in test
    test_indices = shuffled_indices[:test_set_size] # every index up to number of test rows
    train_indices = shuffled_indices[test_set_size:] # the rest of the rows are for training
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
print(len(train_set), "train +", len(test_set), "test")
```

The aim of this is to hide a selection of instances, even across multiple run. What we have done here with a random seed will not work accross multiple runs however. This can be solved if each row, even across runs, has a unique identifyier.

In the book the solution is to compute the *hash* of each instances identifier and keep the last byte of the hash. One would then put the row in the test if it's value is lower than 51 ( ~20% of 256).

Why is this important? - Because we want the test set to remain constant accross multiple runs (*Data Snooping Bias*), even if we refresh the dataset.

Below is one implementation

**Understand what is happening here!**

```python
# The below does not run for Python 2 - for alternatives see original notebook

import hashlib
# What is the hash of an instance?

def test_set_check(identifier, test_ratio, hash):
    # What does digest method do?
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
# Housing doesn't have identifier column -> Use row index
housing_with_id = housing.reset_index() # method in pandas that adds index column
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

### Potential Issues

## Sampling Methods

If this dataset if very large we don't have to be too conserned about our sampling methods. With smaller datasets one can run into sampling bias.

If an expert told us that median income is very important, we may want to ensure our test set is representative of the various categories of income.

```python
housing['median_income'].hist()
```

We'll create an income category attribute by dividing income by 1.5 (to ensure there are many instances per category) and rounding up using `ceil`. Categories greater than 5 will be lumped into 5

```python
housing["income_cat"] = np.ceil(housing["median_income"]/1.5)
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace = True)
```

### Stratified Sampling

Can use Scikit-Learn's `StratifiedShuffleSplit` class.

Basically, it split the dataset according to the *strata* of choice.

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, # Number of re-shuffling & splitting iterations
                               test_size = 0.2, random_state = 42)

# Has the method split, which generates indices to split data into training and test set.
# The second argument is the target variable for supervised learning problems. Stratification is done based on the y labels.
# The split() method returns two outputs train : ndarray - The training set indices for that split.
# and test : ndarray - The testing set indices for that split.

for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

Let's see what proportions of income categories the stratified test set has

```python
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
# Now compare strat_test to overall dist to random test diff

# Define function to output proportions of income categories 
# given dataset 
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

# From earlier, using Scikit's function
train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 42)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set)
}).sort_index()

compare_props["Rand. % error"] = ((compare_props["Random"]-compare_props["Overall"])/compare_props["Overall"])*100
compare_props["Strat. % error"] = ((compare_props["Stratified"]-compare_props["Overall"])/compare_props["Overall"])*100
```

```python
compare_props
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

## Custom Transformers

Will need to write transformers for custom cleanup operations/combining attributes.

Will want transformer to work seamlessly with Scikit-Learn functionalities (like pipelines).

Scikit-Learn relies on duck typing (not inheritance) - *see notes on chapter 2 for into on duck typing*

**Creating a custom transformer**

1. Create a class with 3 methods
    - `fit()` (returning `self`)
    - `transform()`
    - `fit_transform()` You can get the last one for free by adding `TransformerMixin` as a base class
2. If you add `BaseEstimator` as base class *and* aviod `*args` and `**kargs` (functions arguments) in constructor (`__init__`)you'll get two extra methods useful for automatic hyperparameter tuning
    - `get_params()`
    - `set_params()`
    
Below is a small transformer that adds the combined attributes discussed earlier

```python
from sklearn.base import BaseEstimator, TransformerMixin

# Get the right column indices: Safer than hard-coding indices 3,4,5,6 - done in book
# list(housing.columns).index("total_rooms") - gives 3

rooms_ix, bedrooms_ix, population_ix, household_ix = [
    list(housing.columns).index(col)
    for col in ("total_rooms", "total_bedrooms", "population", "households")]

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self # nothing else to do
    def transform(self, X, y = None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_houshold = X[:, population_ix] / X[:, household_ix]
        # Use the class attribute defined earlier
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            # np.c_ is a horizontal concatenation
            return np.c_[X, rooms_per_household, population_per_houshold,
                        bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_houshold]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room = False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

The transformer has one hyperparameter, `add_bedrooms_per_room` set to `True` as default. It is helpful to provide sensible defaults


Alternatively, you can use Scikit-Learn's FunctionTransformer class that lets you easily create a transformer based on a transformation function (thanks to Hanmin Qin for suggesting this code). Note that we need to set validate=False because the data contains non-float values (validate will default to False in Scikit-Learn 0.22).

```python
from sklearn.preprocessing import FunctionTransformer

def add_extra_features(X, add_bedrooms_per_room=True):
    rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
    population_per_household = X[:, population_ix] / X[:, household_ix]
    if add_bedrooms_per_room:
        bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
        return np.c_[X, rooms_per_household, population_per_household,
                     bedrooms_per_room]
    else:
        return np.c_[X, rooms_per_household, population_per_household]

attr_adder = FunctionTransformer(add_extra_features, validate=False,
                                 kw_args={"add_bedrooms_per_room": False})
housing_extra_attribs = attr_adder.fit_transform(housing.values)
```

```python
housing_extra_attribs = pd.DataFrame(housing_extra_attribs,
                                    columns = list(housing.columns)+
                                    ["rooms_per_houshold", 
                                     "population_per_household"])
housing_extra_attribs.head()
```

## Transformation Pipelines

Many data transformations steps need to be executed in the right order. Scikit-Learn provides the `Pipeline` class to help with these sequences. Here's one for the numerical attributes.

Note that we could use CombinedAttributesAdder() instead of FunctionTransformer(...) if we preferred:

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = "median")),
    ('attribs_adder', FunctionTransformer(add_extra_features, validate = False)),
    ('std_scaler', StandardScaler())
    ])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

The pipeline constructor take a list of name/estimator pairs and defines a sequence of steps. The last estimator *must* be a transformer (i.e. must have a `fit_transform()` method)

We can use `ColumnTransformer` to create a pipeline for all transformations (introdced in Scikit-Learn 0.20)

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attribs = list(housing_num) # housing_num = housing.drop("ocean_proximity", axis = 1)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs)
])

housing_prepared = full_pipeline.fit_transform(housing)
```

## Better Evaluation Using Cross-Validation

Can use Scikit-Learn's *cross_validation* feature. 10-fold cv would result in 10 evaluation scores.

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                        scoring = "neg_mean_squared_error", cv = 10) # only MSE option
tree_rmse_scores = np.sqrt(-scores)
```

Scikit-Learn cross-validation features expect a utility function (maximise - greater is better) rather than cost function. Which is why is Neg MSE

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
display_scores(tree_rmse_scores)
```

Perfoms worse than linear regression!

Cross-validation allows one to know how precise the esimate it since one can obtain a standard deviation but at the cost of training the model.

Let's compute the score for linear regression, just to be sure

```python
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
                            scoring = "neg_mean_squared_error", cv = 10)

lin_rmse_scores = np.sqrt( - lin_scores)
display_scores(lin_rmse_scores)
```

**The Decision Tree was overfitting so badly it was worse than linear regression!**

Let's try one last model: `RandomForestRegressor`

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=10, random_state=42)
forest_reg.fit(housing_prepared, housing_labels)
```

```python
# Evaluate scores on the training set
housing_predictions = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, housing_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels,
                                scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

**Training Score** < **Validation Score** = **Overfitting!**

- Can simplify model (regularize it)
- Get more training data

# Fine-Tune Your Model

Once we have our shortlist of promising models we need to fine-tune them. There are several way of doing this.

## Grid Search

Fiddle with hyperparameters manually - very tedius work

Can use Scikit-Learns `GridSearchCV` to search for you, just tell it all the hyperparamers to try out and what values and it will try every combination using cross-validation.

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    # try 12 (3x4) combinations of hyperparameters
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # then we'll try 6 (2x3) combinations with bootstrap set as false
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
]

# previously -> 
# forest_reg = RandomForestRegressor(n_estimators=10, random_state=42)
forest_reg = RandomForestRegressor(random_state = 42)
# Training across 5 fold, total of (12+6)5=90 rounds of training
grid_search = GridSearchCV(forest_reg, param_grid, cv = 5,
                          scoring = 'neg_mean_squared_error',
                          return_train_score = True)
grid_search.fit(housing_prepared, housing_labels)
```

```python
# get best combination of parameters
grid_search.best_params_
```

```python
# try again with larger values
param_grid = [
    # try 12 (3x4) combinations of hyperparameters
    {'n_estimators': [30, 60, 120], 'max_features': [8, 9, 10]},
]

# previously -> 
# forest_reg = RandomForestRegressor(n_estimators=10, random_state=42)
forest_reg = RandomForestRegressor(random_state = 42)
# Training across 5 fold, total of (12+6)5=90 rounds of training
grid_search = GridSearchCV(forest_reg, param_grid, cv = 5,
                          scoring = 'neg_mean_squared_error',
                          return_train_score = True)
grid_search.fit(housing_prepared, housing_labels)
```

```python
# get best combination of parameters
grid_search.best_params_
```

```python
# get best estimator directly
grid_search.best_estimator_
```

If `GridSearchCV` is initialized with `refit=True` then once it finds th best estimator using cv it retrains on the whole training set.

Scores are also available

Let's look at the score of each hyperparameter combination tested during the grid search:

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

Better than model using default parameters! Yay!
